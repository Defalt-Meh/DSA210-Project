# 2.3 Data formatting

```python
spotify_19_24["Play-Time"]= pd.to_datetime(spotify_19_24["ts"]) # To create a additional column
```

```python
spotify_19_24['year'] = pd.DatetimeIndex(spotify_19_24["Play-Time"]).year
spotify_19_24['month'] = pd.DatetimeIndex(spotify_19_24["Play-Time"]).month
spotify_19_24['day'] = pd.DatetimeIndex(spotify_19_24["Play-Time"]).day
spotify_19_24['weekday'] = pd.DatetimeIndex(spotify_19_24["Play-Time"]).weekday
spotify_19_24['time'] = pd.DatetimeIndex(spotify_19_24["Play-Time"]).time
spotify_19_24['hours'] = pd.DatetimeIndex(spotify_19_24["Play-Time"]).hour
spotify_19_24['day-name'] = spotify_19_24["Play-Time"].apply(lambda x: x.day_name())
spotify_19_24['Count'] = 1
```

```python
spotify_19_24["Time-Played (hh-mm-ss)"] = pd.to_timedelta(spotify_19_24["ms_played"], unit='ms')
```

```python
def hours(td):
    # To get the hour information
    return td.seconds/3600

def minutes(td):
    # To get the minutes information
    return (td.seconds/60)%60

spotify_19_24["Listening Time(Hours)"] = spotify_19_24["Time-Played (hh-mm-ss)"].apply(hours).round(3)
spotify_19_24["Listening Time(Minutes)"] = spotify_19_24["Time-Played (hh-mm-ss)"].apply(minutes).round(3)
```

```python
spotify_19_24.head() # To check the newly formed dataset with additional columns
```

```python
spotify_19_24.drop(columns=["ts","Time-Played (hh-mm-ss)","ms_played"], inplace=True) #Drop unnecessary columns
spotify_19_24.describe() # Final check for any abnormality
```

```python
import pandas as pd
import folium

# Load the dataset (replace with the actual path to your file if needed)
# df = pd.read_csv('Spotify_19_24.csv')  # Uncomment if loading from a CSV file
df = spotify_19_24  # Use your DataFrame variable

# 1. Find the most listened-to song in each country
most_listened = df.groupby('conn_country').apply(
    lambda group: group.loc[group['count'].idxmax()]
).reset_index(drop=True)

# 2. Add coordinates for each country (predefined or using a geocoding API)
country_coords = {
    'TR': {'lat': 38.9637, 'lon': 35.2433},  # Turkey
    'DE': {'lat': 51.1657, 'lon': 10.4515},  # Germany
    'GB': {'lat': 55.3781, 'lon': -3.4360}   # United Kingdom
}

# Add latitude and longitude columns to the DataFrame
most_listened['latitude'] = most_listened['conn_country'].map(lambda x: country_coords.get(x, {}).get('lat', None))
most_listened['longitude'] = most_listened['conn_country'].map(lambda x: country_coords.get(x, {}).get('lon', None))

# 3. Create a map using folium
m = folium.Map(location=[20, 0], zoom_start=2)  # Center the map globally

# Add markers for each country
for _, row in most_listened.iterrows():
    if pd.notnull(row['latitude']) and pd.notnull(row['longitude']):  # Only add marker if coordinates are available
        popup_text = (
            f"<b>Country:</b> {row['conn_country']}<br>"
            f"<b>Most Listened Song:</b> {row['master_metadata_track_name']}<br>"
            f"<b>Play Count:</b> {row['count']}"
        )
        folium.Marker(
            location=[row['latitude'], row['longitude']],
            popup=folium.Popup(popup_text, max_width=300),
            tooltip=row['conn_country']
        ).add_to(m)

# Save the map to an HTML file (optional)
m.save("spotify_most_listened_map.html")

# Display the map
m
```
